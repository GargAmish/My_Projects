```python
#Data analysis
#calculating unique values of a column
cancelled_perc= df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
#calculating percentage by hotel
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Visualisation
#bar graph
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
#count graph
plt.figure(figsize=(8,4))
axes=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues') #creating count plot
plt.show()
```

```python
#grouping by price
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#plotting graph by date and price mean
plt.figure(figsize=(20,8))
plt.title('Average Daile Rate by Hotels',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
axes=sns.countplot(x=df['month'],hue='is_canceled',data=df,palette='bright')
plt.show()
```

```python
#plotting adr by month
plt.figure(figsize=(15,8))
plt.title('ADR by Month',fontsize=30)
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
#pie plot by countries
cancelled=df[df['is_canceled']==1]
top_10_countries=cancelled['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries with Cancelled Reservations',fontsize=25)
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)
plt.show()
```
